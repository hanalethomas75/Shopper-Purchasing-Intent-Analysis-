```python
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/hanalethomas75/My-Awesome-Repository/refs/heads/main/online_shoppers_intention.csv')
df
```

```python
#df,info helps structure the data types
df.info()
#df,describe summarizes the statistics
df.describe()
#df.isnull.sum checks for missing values
df.isnull().sum()
#df.duplicated checks for any duplicated values
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
df.duplicated().sum()
```

```python
#correlation Analysis
import matplotlib.pyplot as plt
import seaborn as sns

corr_matrix = df.corr(numeric_only=True)

plt.figure(figsize=(12,18))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.boxplot(x=df['Revenue'], y=df['PageValues'])
plt.title('Page Values by Revenue')
plt.xlabel('Revenue')
plt.ylabel('Page Values')

# Interpretation:
# Purchasing sessions tend to have higher PageValues,
# though there is overlap between the groups.
plt.show()
```

```python
plt.figure(figsize=(10,8))
df_plot = df.copy()
sns.boxplot(x=df['Revenue'], y=df['BounceRates'])
plt.title('Bounce Rates by Revenue')
plt.xlabel('Revenue')
plt.ylabel('Bounce Rates')

# Interpretation:
# Users who generate revenue tend to have lower bounce rates.
# High bounce rates indicate users leaving quickly without purchasing.
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.histplot(df['ProductRelated_Duration'], kde=True)
plt.title('Distribution of Product Related Duration')
plt.xlabel('Duration')
plt.ylabel('Frequency')

# Interpretation:
# Most users spend low time on product pages, with a long right tail.
# A few users spend significantly more time, which may relate to purchases.
plt.show()
```

```python
plt.figure(figsize=(5,4))
sns.countplot(x=df['Revenue'])
plt.title('Revenue Distribution')
plt.xlabel('Revenue')
plt.ylabel('Count')

# Interpretation:
# The dataset is imbalanced, with more non-purchasing sessions than purchasing ones.
plt.show()
```
